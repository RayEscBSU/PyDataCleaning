```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\rescobedo\Documents\test\Ada County (003).xls",sheet_name='Sheet2')
```

```python
columsToDrop = ['Unnamed: 6', 'Unnamed: 13', 'Unnamed: 18', 'Unnamed: 20', 'Unnamed: 22']
```

```python
df= df.drop(columns=columsToDrop) #dro unused colums
```

```python
df = df.fillna('')
```

```python
# Trim leading and trailing whitespaces from all string values
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#drop top tow rows
df = df.iloc[2:] # delete top rows
```
